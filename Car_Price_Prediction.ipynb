```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
y = train.iloc[:,10:11]
y
```

```python
x = train.drop(['SalePrice'],axis=1)
x
```

```python
Y = test.iloc[:,10:11]
X = test.drop(['SalePrice'],axis=1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
y_pred = lr.predict(X)
```

```python
from sklearn.metrics import r2_score
print(f"Araba fiyat tahmini doÄŸruluk oranÄ± {r2_score(Y,y_pred)}")
```

```python
from sklearn.metrics import mean_squared_error
print(f"Araba fiyat tahmini mean_squared_error {mean_squared_error(Y,y_pred)}")
```
